# SF Crime data exploration

## 1. Load the data

```python
import pandas as pd
from zipfile import ZipFile

sf_crime_zip = ZipFile('data/sf-crime.zip')
df = pd.read_csv(sf_crime_zip.open('train.csv'))
df_test = pd.read_csv(sf_crime_zip.open('test.csv'))
df_submit = pd.read_csv(sf_crime_zip.open('sampleSubmission.csv'))
```
